# Here, we will be importing several python libraries.


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

# What kind of coffee sells the most?

```python
sales_by_coffee = df.groupby('coffee_name')['money'].sum().sort_values(ascending=False)
sales_by_coffee.plot(kind='bar', figsize=(10,5))
plt.title('Total Sales by Coffee Type')
plt.xlabel('Coffee Type')
plt.ylabel('Total Money ($)')
plt.show()
```

```python
sns.heatmap(df.select_dtypes(include=['number']).corr(), annot=True)
plt.title("Correlation Heatmap (Numeric Columns Only)")
plt.show()
```
