```python
import pandas as pd
df = pd.read_csv("/Users/rainnebus/Downloads/archive/index_1.csv")
df.head()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
sales_by_coffee = df.groupby('coffee_name')['money'].sum().sort_values(ascending=False)
sales_by_coffee.plot(kind='bar', figsize=(10,5))
plt.title('Total Sales by Coffee Type')
plt.xlabel('Coffee Type')
plt.ylabel('Total Money ($)')
plt.show()
```
